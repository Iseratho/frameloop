```python
import string

label_space = list(string.ascii_uppercase[:10])
label_space, len(label_space)
```

```python
import numpy as np
example_list = ["A", "B", "A", "A", "A", "B", "C", "A"]

imbalanced  = ["A"]+["B"]*999
farapart = ["A"]+["B"]*998+["A"]
allsame= ["A"]*1000
alldiff = label_space
cycle = label_space*100
rand = list(map(lambda x: label_space[x], np.random.randint(10, size=1000_000)))
alternate = ["A", "B"]*500
```

```python
import itertools

def drr(lst):
    def num_direct_repeat_events(lst):
        pairs = itertools.pairwise(lst)
        same_pairs = map(lambda x: x[0] == x[1], pairs)
        return sum(same_pairs)
    if len(lst) < 2:
        return 0
    num = num_direct_repeat_events(lst)
    return num/(len(lst) - 1)
```

```python
drr(allsame), drr(imbalanced), drr(farapart), drr(example_list), drr(rand), drr(alternate), drr(cycle), drr(alldiff), drr([])
```

```python
from collections import defaultdict

def rrdist(lst):
    def count_items_between_duplicates(lst):
        counts = defaultdict(int)
        index_dict = {}

        for i, item in enumerate(lst):
            if item in index_dict:
                distance = i - index_dict[item]
                counts[distance] += 1
            index_dict[item] = i

        return counts
    cnt = count_items_between_duplicates(lst)
    if not cnt:
        return 0
#     return 1/((sum(k*v for k,v in cnt.items()))/(sum(cnt.values())))=
    return sum(cnt.values())/sum(k*v for k,v in cnt.items())
```

```python
rrdist(allsame), rrdist(imbalanced), rrdist(alternate), rrdist(farapart), rrdist(example_list), rrdist(rand), rrdist(cycle), rrdist(alldiff), rrdist([])
```

```python
def uniq(lst, labels=label_space):
    if len(lst) < 2:
        return 1
    max_uniq = min(len(lst), len(labels))-1
    uniq_in_list = len(set(lst))-1
    return uniq_in_list/max_uniq
```

```python
uniq(allsame), uniq(imbalanced), uniq(farapart), uniq(alternate), uniq(example_list), uniq(rand), uniq(cycle), uniq(alldiff), uniq([])
```

```python
from collections import Counter

# https://www.kdnuggets.com/2020/01/decision-tree-algorithm-explained.html#:~:text=Gini%20Index%20works%20with%20the,implies%20higher%20inequality%2C%20higher%20heterogeneity.&text=Calculate%20Gini%20for%20sub%2Dnodes,)%20(p%C2%B2%2Bq%C2%B2).
def gini(lst):
    counts = Counter(lst)
    sum_probs = sum((i/len(lst))**2 for _, i in counts.items())
    return 1 - sum_probs
```

```python
gini(allsame), gini(imbalanced), gini(farapart), gini(alternate), gini(example_list), gini(rand), gini(cycle), gini(alldiff), gini([])
```
