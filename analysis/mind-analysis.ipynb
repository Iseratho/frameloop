# Behavior Log Analysis

```python
behavior_df = pd.read_csv("../input/mind-news-dataset/MINDsmall_train/behaviors.tsv", sep="\t", header=None, index_col=0, names=["user_id", "datetime", "history", "impression"])
behavior_df["history"] = behavior_df["history"].fillna("")
behavior_df
```

```python
(behavior_df[['user_id','datetime']].value_counts() > 1).sum()
```

```python
behavior_df[behavior_df["datetime"] == "11/10/2019 9:08:07 AM"]
```

```python
behavior_df.loc[125629]["impression"]
```

```python
behavior_df.describe()
```

```python
def count_clicks(x):
    imps = [int(i.split("-")[1]) for i in x.split()]
    return sum(imps)

behavior_df["impression"].map(count_clicks).describe()
```

```python
behavior_df["impression"].map(count_clicks).sum()
```

```python
behavior_df[behavior_df["impression"].map(count_clicks) >= 2]
```

```python
behavior_df["impression"].str.split().map(len).describe()
```

```python
behavior_df.drop_duplicates(subset="user_id")["history"].str.split().map(len).describe()
```
