```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("/kaggle/input/frameloop-result-table/table.csv").set_index(["seq","metric"])
df
```

```python
base_seq = "UserHistory"
for ltype in ["cat", "media", "moral", "sem"]:
    for mtype in ["𝐷𝑅𝑅", "𝑅𝑅𝑑𝑖𝑠𝑡", "𝑈𝑛𝑖𝑞", "𝐺𝑖𝑛𝑖"]:
        for seqtype in ["PerCategory", "WithImpressions", "WithClicks"]: 
            base_row = df.loc[base_seq, mtype]
            base_stats = (float(base_row[ltype + "_mean"]), float(base_row[ltype + "_std"]), base_row["num"])
            comp_row = df.loc[seqtype, mtype]
            if "-" == comp_row[ltype + "_mean"]:
                continue
            comp_stats = (float(comp_row[ltype + "_mean"]), float(comp_row[ltype + "_std"]), comp_row["num"])
            res = stats.ttest_ind_from_stats(*base_stats, *comp_stats)
            symb = "✓" if res[1] < 0.0005 else "❌"
            print(base_seq, seqtype, mtype, ltype, res, symb)
```
